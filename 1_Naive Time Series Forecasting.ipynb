```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Data

```python
df = pd.read_csv("data/daily_min_temp.csv", index_col='Date',parse_dates=["Date"])
df.head()
```

# Resampling Data Time Series

```python
df.Temp.plot(figsize=(15,5))
```

```python
df = df.resample('m').mean()
df
```

```python
df.Temp.plot(figsize=(15,5))
```

# Naive Forecasting

## One-shif

```python
pred = df.copy()
pred = pred.shift()
pred
```

```python
plt.figure(figsize=(15,5))
plt.plot(df,'b-')
plt.plot(pred,'r-')
```

## Mean as Forecast

```python
pred =df.copy()
pred.Temp = [df.loc[:date].Temp.mean() for date in df.index]
pred = pred.shift()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df,'b-')
plt.plot(pred,'r-')
```

## Moving Average / Rolling Mean Forecast

```python
pred = df.copy()
pred.Temp = df.rolling(3).mean()
pred = pred.shift()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df,'b-')
plt.plot(pred,'r-')
```
